```python
import pandas as pd
import xml.etree.ElementTree as ET
import glob
```

### Function definition to convert XML file into a DataFrame

```python
def convertXmlToDf(year_dir, state_dir, xml_file_path, file_counter):
    xml_df = pd.DataFrame()
    counter = 1
    with open(xml_file_path, 'r') as xml_file:
        file_content = xml_file.read()
        etree = ET.fromstring(file_content) #create an ElementTree object
        doc_elements = list(etree[1])
        for doc in doc_elements:
            data_point_elements = list(doc)
            data_point_dict = {}
            for data_point in data_point_elements:
                dp_tagname = data_point.tag[37:]
                dp_tagval = data_point.text
                data_point_dict[dp_tagname] = dp_tagval
            temp_df = pd.DataFrame(data_point_dict, index = ['{}_{}_{}_{}'.format(year_dir, state_dir, file_counter, counter),])
            counter = counter + 1
            if xml_df.empty:
                xml_df = temp_df
            else:
                frames = [xml_df, temp_df]
                xml_df = pd.concat(frames)
    return xml_df
```
