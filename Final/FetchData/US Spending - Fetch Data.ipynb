```python
import requests
import os
import json
import time
import string
import xml.etree.ElementTree as ET
```

```python
all_state_codes = ['MA','PA']
all_contract_years = ['2016','2017']
contract_url = "https://www.usaspending.gov/fpds/fpds.php"
max_records = 100
```

### Download and save contracts in xml files

```python
def firstApiCall(state_code, contract_year, reqparams, contract_url):
    # Send HTTP request
    r = requests.get(contract_url, params = reqparams)
    # Parse the response into an xml
    xml_root = ET.fromstring(r.text)
    # Find total number of records present
    numFound = xml_root[1].attrib.get('numFound')
    # Download and save the xml for first iteration
    data_save_path = '../Data/' + contract_year + '/' + state_code + '/' + '0.xml'
    data_save_dirname = os.path.dirname(data_save_path)
    os.makedirs(data_save_dirname,exist_ok=True)
    with open(data_save_path, 'w') as doc_file:
        doc_file.write(r.text)
    return numFound
```

### Determine the number of iterations to download  and save xml files each having 100 records

```python
def subsequentApiCalls(state_code, contract_year, reqparams, contract_url, numFound, max_records):
    # Calculating the number of iterations
    loop_counter_max = int(numFound)/max_records
    loop_counter_max = int(loop_counter_max)
    # Iterating to change the request parameters to get the next max_records(for example 100) records
    for x in range(1,loop_counter_max):
        reqparams['records_from'] = reqparams['records_from'] + max_records
        data_save_path = '../Data/' + contract_year + '/' + state_code + '/' + str(x) + '.xml'
        if not os.path.exists(data_save_path):
            r = requests.get(contract_url, params = reqparams)
            if r.status_code == 200:
                try:
                    xml_root = ET.fromstring(r.text)
                    with open(data_save_path, 'w') as doc_file:
                        doc_file.write(r.text)
                except:
                    print('Following parse error occured:')
                    print(r.text)
            else:
                print(r.status_code)
                print('Download failure for', state_code, contract_year, x)
        else:
            print('File already exists', x)
```

```python
for contract_year in all_contract_years:
    for state_code in all_state_codes:
        #Request Map for contracts data
        reqparams = {'detail':'b', 'fiscal_year':contract_year,'stateCode':state_code, 'max_records':max_records, 'records_from':0}
        numFound = firstApiCall(state_code, contract_year, reqparams, contract_url)
        print('First API call completed for', contract_year, state_code)
        subsequentApiCalls(state_code, contract_year, reqparams, contract_url, numFound, max_records)
        print('Subsequent API calls completed for', contract_year, state_code)
```
