## Q8 Use nps_chat and wordlist corpus to find unusual words used by people in chats

```python
from nltk.corpus import nps_chat
from nltk.corpus import words
```

```python
wordList = words.words()
uniqueChatWords = set([word.lower() for sentence in nps_chat.posts() for word in sentence])
validWordsLower = set([word.lower() for word in wordList])
chatOnlyWords = uniqueChatWords - validWordsLower
print('Total number of chat only words which are not in words corpus is:', len(chatOnlyWords))
print('\nThese words are:\n\n', chatOnlyWords)
```
